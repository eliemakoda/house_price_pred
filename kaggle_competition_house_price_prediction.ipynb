```python
#let's import the tools we'll be using
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
#read the csv train and test file
train_dataset= pd.read_csv("train.csv")
test_dataset = pd.read_csv("/content/test.csv")
```

```python
#let's try to one hot encode our non numerical data
#you can use the reverse_tranform method to view the encoded label
#instanciate the label encoder
le= LabelEncoder()
def encode_dataset(train_dataset=train_dataset):
  """
  this function uses the fit transform method to encode non numerical data in our dataset

  params:
      train_dataset: the target dataset
  return :
       returns a new encoded dataset
  """
  train_dataset["MSZoning"]= le.fit_transform(train_dataset["MSZoning"])
  train_dataset["Street"]=le.fit_transform(train_dataset["Street"])
  train_dataset["Alley"]=le.fit_transform(train_dataset["Alley"])
  train_dataset["LotShape"]=le.fit_transform(train_dataset["LotShape"])
  train_dataset["LandContour"]=le.fit_transform(train_dataset["LandContour"])
  train_dataset["Utilities"]=le.fit_transform(train_dataset["Utilities"])
  train_dataset["LotConfig"]=le.fit_transform(train_dataset["LotConfig"])
  train_dataset["LandSlope"]=le.fit_transform(train_dataset["LandSlope"])
  train_dataset["Neighborhood"]=le.fit_transform(train_dataset["Neighborhood"])
  train_dataset["Condition1"]=le.fit_transform(train_dataset["Condition1"])
  train_dataset["Condition2"]=le.fit_transform(train_dataset["Condition2"])
  train_dataset["BldgType"]=le.fit_transform(train_dataset["BldgType"])
  train_dataset["HouseStyle"]=le.fit_transform(train_dataset["HouseStyle"])
  train_dataset["RoofStyle"]=le.fit_transform(train_dataset["RoofStyle"])
  train_dataset["RoofMatl"]=le.fit_transform(train_dataset["RoofMatl"])
  train_dataset["Exterior1st"]=le.fit_transform(train_dataset["Exterior1st"])
  train_dataset["Exterior2nd"]=le.fit_transform(train_dataset["Exterior2nd"])
  train_dataset["MasVnrType"]=le.fit_transform(train_dataset["MasVnrType"])
  train_dataset["ExterQual"]=le.fit_transform(train_dataset["ExterQual"])
  train_dataset["ExterCond"]=le.fit_transform(train_dataset["ExterCond"])
  train_dataset["Foundation"]=le.fit_transform(train_dataset["Foundation"])
  train_dataset["BsmtQual"]=le.fit_transform(train_dataset["BsmtQual"])
  train_dataset["BsmtCond"]=le.fit_transform(train_dataset["BsmtCond"])
  train_dataset["BsmtExposure"]=le.fit_transform(train_dataset["BsmtExposure"])
  train_dataset["BsmtFinType1"]=le.fit_transform(train_dataset["BsmtFinType1"])
  train_dataset["BsmtFinType2"]=le.fit_transform(train_dataset["BsmtFinType2"])
  train_dataset["Heating"]=le.fit_transform(train_dataset["Heating"])
  train_dataset["HeatingQC"]=le.fit_transform(train_dataset["HeatingQC"])
  train_dataset["CentralAir"]=le.fit_transform(train_dataset["CentralAir"])
  train_dataset["Electrical"]=le.fit_transform(train_dataset["Electrical"])
  train_dataset["KitchenQual"]=le.fit_transform(train_dataset["KitchenQual"])
  train_dataset["Functional"]=le.fit_transform(train_dataset["Functional"])
  train_dataset["FireplaceQu"]=le.fit_transform(train_dataset["FireplaceQu"])
  train_dataset["GarageType"]=le.fit_transform(train_dataset["GarageType"])
  train_dataset["GarageFinish"]=le.fit_transform(train_dataset["GarageFinish"])
  train_dataset["GarageQual"]=le.fit_transform(train_dataset["GarageQual"])
  train_dataset["GarageCond"]=le.fit_transform(train_dataset["GarageCond"])
  train_dataset["PoolQC"]=le.fit_transform(train_dataset["PoolQC"])
  train_dataset["Fence"]=le.fit_transform(train_dataset["Fence"])
  train_dataset["MiscFeature"]=le.fit_transform(train_dataset["MiscFeature"])
  train_dataset["SaleType"]=le.fit_transform(train_dataset["SaleType"])
  train_dataset["SaleCondition"]=le.fit_transform(train_dataset["SaleCondition"])
  train_dataset["PavedDrive"]=le.fit_transform(train_dataset["PavedDrive"])
  train_dataset["LotFrontage"]=le.fit_transform(train_dataset["LotFrontage"])
  train_dataset["MasVnrArea"]=le.fit_transform(train_dataset["MasVnrArea"])
  train_dataset["GarageYrBlt"]=le.fit_transform(train_dataset["GarageYrBlt"])
  return train_dataset
```

```python
#encode our data using our function
train_dataset= encode_dataset(train_dataset)
train_dataset.head(3)
```

```python
#remove the id column from our dataset
train_dataset.drop("Id",axis=1, inplace=True)
```

```python
#check if there's any missing NAN values in our dataset
#test_dataset.isna().any()[lambda x: x]
len(test_dataset.columns),len(train_dataset.columns)
```

```python
#devide our data in x and y
X= train_dataset.drop("SalePrice",axis=1)
Y=train_dataset["SalePrice"]
```

```python
#split our data (devide using the train test_split method )
xtrain,xtest,ytrain,ytest= train_test_split(X,Y, test_size=0.2)
#instanciate our model
model= LinearRegression()
#fit our model
model.fit(xtrain,ytrain)
#evaluate
model.score(xtest,ytest)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
reg = make_pipeline(StandardScaler(),
                    SGDRegressor())
reg.fit(xtrain,ytrain)
reg.score(xtest,ytest)
```

```python
from sklearn import linear_model
scaler= StandardScaler()
regres= linear_model.BayesianRidge()
regres.fit(xtrain,ytrain)
regres.score(xtest,ytest)
```

```python
#test another model
from sklearn.linear_model import ElasticNet
regr = ElasticNet(random_state=0,max_iter=1000)
regr.fit(xtrain,ytrain)
regr.score(xtest,ytest)
```

```python
from sklearn import linear_model
clf = linear_model.Ridge(alpha=0.1)
clf.fit(xtrain,ytrain)
clf.score(xtest,ytest)
```
